## Prosper Loan Data Exploratory Analysis

Import Packages:

```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline
```

Read & Analyse Data assets:

```python
df = pd.read_csv("prosperLoanData.csv")
```

```python
df.info()
```

```python
df.describe()
```

#### Univariate Analysis:

In this step, we will be study Individual Variables.

```python
plt.figure(figsize = [10, 5])
order=df['LoanStatus'].value_counts()
sb.countplot(data=df,x="LoanStatus",order=order.index)
plt.xticks(rotation=90);
```

Observations:
Looking at the Loan Status within Prosper Data, its visible that the majority of the Loan status is Active followed by Completed.


```python
order=df['Term'].value_counts()
plt.pie(order, labels = order.index, startangle = 90,counterclock = False,  rotatelabels =True);
plt.axis('square')

plt.title('Loan Term in Months',loc='left');
```

Observations: It is noted from Prosper Loan data that most Loans are Termed 36 Months, followed by 60 Months

```python

sb.countplot(data=df,x="ProsperRating (numeric)")
plt.xticks(rotation=90);
```

The  Prosper Rating assigned at the time the listing was created as: 0 - N/A, 1 - HR, 2 - E, 3 - D, 4 - C, 5 - B, 6 - A, 7 - AA. The Rating are  Applicable for loans originated after July 2009.
The Class C gets the hightest rating of all, followed by B & A.

```python
plt.figure(figsize = [10, 5])
order=df['ListingCategory (numeric)'].value_counts()
sb.countplot(data=df,x="ListingCategory (numeric)",order=order.index)
plt.xticks(rotation=90);
```

The listing category defined are: 0 - Not Available, 1 - Debt Consolidation, 2 - Home Improvement, 3 - Business, 4 - Personal Loan, 5 - Student Use, 6 - Auto, 7- Other, 8 - Baby&Adoption, 9 - Boat, 10 - Cosmetic Procedure, 11 - Engagement Ring, 12 - Green Loans, 13 - Household Expenses, 14 - Large Purchases, 15 - Medical/Dental, 16 - Motorcycle, 17 - RV, 18 - Taxes, 19 - Vacation, 20 - Wedding Loans.

The Bar chart denotes that the most Borrower apply loans for Debt Consolidation, which usually is to pay of other debts.

```python

sb.countplot(data=df,x="IsBorrowerHomeowner") 
```

A Borrower will be classified as a homowner if they have a mortgage on their credit profile or provide documentation confirming they are a homeowner.

Based on the plot there are almost equal number of Home Owner borrower  and non Home Owner Borrower.

```python
order=df['IncomeRange'].value_counts()
sb.countplot(data=df,x="IncomeRange",order=order.index)

plt.title('IncomeRange')
plt.xticks(rotation=45);
```

Above plot depicts that most borrowers are under the income class of 25k-75k

#### Bivariate Analysis

Analyse IsBorrowerHomeowner and Ratings

```python
plt.figure(figsize = [10, 5])
base_color = sb.color_palette()[0]

# left plot: violin plot
plt.subplot(1, 2, 1);
ax1 = sb.violinplot(data = df, x = 'IsBorrowerHomeowner', y = 'ProsperRating (numeric)', color = base_color);

# right plot: box plot
plt.subplot(1, 2, 2);
sb.boxplot(data = df, x = 'IsBorrowerHomeowner', y = 'ProsperRating (numeric)', color = base_color);
plt.ylim(ax1.get_ylim()) # set y-axis limits to be same as left plot
```

```python
df['Recommendations'].value_counts()
```

The Rating distribution between homeowner and non homeowner is not the same considering that the median to be the same.

```python

# plots the hist2d graph 
plt.hist2d(data=df, x='BorrowerRate', y='LoanOriginalAmount', cmap = 'viridis', cmin = 0.5);
plt.colorbar()

plt.xlabel('BorrowerRate')
plt.ylabel('LoanOriginalAmount')
plt.title('LoanOriginalAmount vs. BorrowerRate')
```

Observations: It is noticed that the rate of the loan is high when the Loan amount borrowed is lower.

```python
order = ['Not employed', '$0', '$1-24,999', '$25,000-49,999', '$50,000-74,999', '$75,000-99,999', '$100,000+']
sb.boxplot(data=df, x='IncomeRange', y='ProsperPrincipalBorrowed', order=order);
plt.yscale('log')
plt.xticks(rotation=90);
plt.ylabel('Principal Borrowed')
```

The above boxplot depicts that as the Income range increases, the more is the principal amount

```python
df['year'] = df.ListingCreationDate.str.split('-').str[0]
sb.pointplot(data=df, x='year', y='BorrowerRate',color =sb.color_palette()[5]);
```

The above graph shows that rate of interest rate increased from 2005 to 2006 and then dropped off from 2006-2007 and then it constantly increated for few years, reached its peak in 2011 which then started decreasing until 2014.

##### Multivariate Aanlysis:

```python

plt.figure(figsize = [10, 5])

g = sb.FacetGrid(data = df, col = 'IsBorrowerHomeowner', size = 6,margin_titles = True)
g.map(sb.boxplot, 'year', 'BorrowerRate')
```

The above plot analyses how the rate of interest changed over years for Homeowners and Non-Homeowners. This boxplot shows that the borrower rate increased for some years in case of non homeowners, however it constantly decreased around 2010. As regards of homeowners, there wasn't much difference till 2009 but after 2009, the graph went up till 2011 and dropped from then.
